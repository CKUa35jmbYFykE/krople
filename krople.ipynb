```python
import pygame
import random

# Inicjalizacja Pygame
pygame.init()

# Ustawienia ekranu
SCREEN_WIDTH = 800
SCREEN_HEIGHT = 600
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Animacja kropli deszczu")

# Kolory
WHITE = (255, 255, 255)
BLUE = (0, 0, 255)

# Parametry kropli deszczu
DROP_SIZE = 3
DROP_SPEED = 5

# Klasa reprezentująca kroplę deszczu
class Raindrop:
    def __init__(self):
        self.x = random.randint(0, SCREEN_WIDTH)
        self.y = random.randint(-SCREEN_HEIGHT, 0)
    
    def fall(self):
        self.y += DROP_SPEED
    
    def draw(self):
        pygame.draw.circle(screen, BLUE, (self.x, self.y), DROP_SIZE)

# Lista kropli deszczu
raindrops = []

# Główna pętla gry
running = True
while running:
    screen.fill(WHITE)
    
    # Obsługa zdarzeń
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    # Ruch i rysowanie kropli deszczu
    for drop in raindrops:
        drop.fall()
        drop.draw()
    
    # Dodawanie nowych kropli w nieskończoność
    if random.random() < 0.05:  # Szansa na dodanie nowej kropli
        raindrops.append(Raindrop())
    
    # Usuwanie kropli, które opuściły ekran
    raindrops = [drop for drop in raindrops if drop.y < SCREEN_HEIGHT]
    
    pygame.display.flip()
    pygame.time.delay(30)

# Zamknięcie Pygame
pygame.quit()
```
